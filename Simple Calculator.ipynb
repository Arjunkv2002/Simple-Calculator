```
def add(x,y):
    a=x+y
    return a
def sub(x,y):
    b=x-y
    return b
def mul(x,y):
    c=x*y
    return c
def div(x,y):
    d=x/y
    return d
x=int(input("ENTER THE FIRST NUMBER :"))
y=int(input("ENTER THE SECOND NUMBER :"))
print("MAIN MENU")
print(" 1) ADDITION \n 2) SUBTRACTION \n 3) MULTIPLICATION \n 4) DIVISION")
ch=int(input(" ENTER YOUR CHOICE FROM ABOVE : "))
if (ch==1):
    e=add(x,y)
    print(" SUM = ",e)
elif (ch==2):
    e=sub(x,y)
    print(" DIFFERENCE = ",e)
elif (ch==3):
    e=mul(x,y)
    print(" PRODUCT = ",e)
elif (ch==4):
    e=div(x,y)
    print(" QUOTIENT = ",e)
else:
    print("INVALID DATA")
```
